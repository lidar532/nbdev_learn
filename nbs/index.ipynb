# nbdev_learn

> Learn how to use NBdev to generate python packages and documentation.

## What is nbdev_learn

`nbdev_learn` is a simple Python Notebook github based project intended as a 
simple example of how to use the `nbdev` tools to generate python packages,
and modules complete with online web based documentation.  It consists of
two module files, `libA` and `libB` each containing two very simple functions.

They are organized as:
- nbdev_learn
  - libA.py
    - func_a()
    - func_b()
  - libB.py
    - func_d()
    - func_e()
   
`libA.py` is generated from the source python notebook `00_libA.ipynb` and `libB.py` is 
generated by `10_libB.ipynb`.
